# LAB 10 - Decision Tree Classifier

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
import seaborn as sns
df = sns.load_dataset('iris')
df.head()
```

```python
X=df.iloc[:,:-1]
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
treemodel=DecisionTreeClassifier()
treemodel.fit(X_train,y_train)
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
# prepruning
treemodel=DecisionTreeClassifier(max_depth=2)
treemodel.fit(X_train,y_train)
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
y_pred=treemodel.predict(X_test)
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(y_test,y_pred)
print(score)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Testing accuracy vs depth of tree
# pruning
treemodel=DecisionTreeClassifier(max_depth=2)
treemodel.fit(X_train,y_train)
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
test_accuracy = []
for depth in range(1, 11):
    treemodel = DecisionTreeClassifier(max_depth=depth)
    treemodel.fit(X_train, y_train)
    y_pred = treemodel.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    test_accuracy.append(score)

# Plotting the accuracy vs depth
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), test_accuracy, marker='o')
plt.xlabel('Depth of Decision Tree')
plt.ylabel('Test Accuracy')
plt.title('Accuracy vs Depth of Decision Tree')
plt.grid(True)
plt.show()
```
